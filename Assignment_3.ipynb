###2. Explain the concept of function arguments and parameters in Python.
Ans.
In Python, parameters are the variables listed in the function definition, while arguments are the actual values passed to the function when it's called.
There are several types of arguments in Python:
a. Positional arguments: Passed in the order defined in the function.
b. Keyword arguments: Passed with parameter names.
c. Default arguments: Have preset values if not provided.
d. Variable-length arguments (*args and **kwargs): Allow passing multiple arguments.

```python
def describe_person(name, age, *hobbies, **traits):
    print(f"{name} is {age} years old.")
    if hobbies:
        print(f"Hobbies: {', '.join(hobbies)}")
    for trait, value in traits.items():
        print(f"{trait.capitalize()}: {value}")

describe_person("Alice", 30, "reading", "painting", hair="brown", eyes="blue")
```

###3. What are the different ways to define and call a function in Python?
There are several ways to define and call functions in Python.

```python
#a. Regular function definition:
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))  # Call

#b. Lambda function (anonymous function):
greet = lambda name: f"Hello, {name}!"
print(greet("Bob"))  # Call

#c. Nested function:
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function

add_five = outer_function(5)
print(add_five(3))  # Call

#d. Method in a class:
class Greeter:
    def greet(self, name):
        return f"Hello, {name}!"

g = Greeter()
print(g.greet("Charlie"))  # Call

```

###4. What is the purpose of the `return` statement in a Python function?

Ans.
The return statement in Python serves two main purposes:

1. It specifies the value that the function should output.
2. It immediately terminates the function execution and returns control to the calling code.

```python
def divide(a, b):
    if b == 0:
        return "Error: Division by zero"
    return a / b

print(divide(10, 2))  # Output: 5.0
print(divide(10, 0))  # Output: Error: Division by zero
```

###5. What are iterators in Python and how do they differ from iterables?
Ans.
Iterators are objects that can be iterated (looped) upon. They implement the __iter__() and __next__() methods. Iterables, on the other hand, are objects that can return an iterator.
The main difference is that an iterator keeps track of its current state, while an iterable does not.

```python
# Iterable
my_list = [1, 2, 3]  # List is an iterable

# Iterator
my_iterator = iter(my_list)  # Creating an iterator from the list

print(next(my_iterator))  # Output: 1
print(next(my_iterator))  # Output: 2
print(next(my_iterator))  # Output: 3
```

###6. Explain the concept of generators in Python and how they are defined.
Ans.
Generators are a type of iterable, like lists or tuples. However, unlike lists, they don't store all the values in memory, but generate the values on the fly. They are created using a function with a yield statement instead of return.

```python
def countdown(n):
    while n > 0:
        yield n
        n -= 1

for number in countdown(5):
    print(number)
```

###7. What are the advantages of using generators over regular functions?
Ans.
Generators have several advantages:
a. Memory Efficiency: They generate values on-the-fly, not storing all values in memory.
b. Performance: Can start producing values immediately, without waiting for all processing.
c. Infinite Sequences: Can represent infinite sequences, which is impossible with normal functions.
d. Simplicity: Often lead to simpler code for complex iterations.

```python
#Example comparing a generator to a regular function:
# Generator
def even_numbers_gen(n):
    for i in range(n):
        if i % 2 == 0:
            yield i

# Regular function
def even_numbers_list(n):
    return [i for i in range(n) if i % 2 == 0]

# Generator usage
for num in even_numbers_gen(1000000):
    if num > 10:
        break
    print(num)

# List usage (less efficient for large n)
even_list = even_numbers_list(1000000)
for num in even_list:
    if num > 10:
        break
    print(num)
```

###9. Explain the purpose and usage of the map() function in Python.
Ans.
The map() function applies a given function to each item of an iterable (like a list) and returns a map object (which can be converted to a list or iterated over).
Purpose: To transform all elements in an iterable without using an explicit for loop.

```python
def celsius_to_fahrenheit(c):
    return (c * 9/5) + 32

temperatures_c = [0, 10, 20, 30, 40]
temperatures_f = list(map(celsius_to_fahrenheit, temperatures_c))
print(temperatures_f)  # Output: [32.0, 50.0, 68.0, 86.0, 104.0]
```

###10. What is the difference between map(), reduce(), and filter() functions in Python?

Ans.
map(): Applies a function to all items in an input list.
filter(): Creates a list of elements for which a function returns true.
reduce(): Applies a rolling computation to sequential pairs of values in a list.

```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

# map() - Double each number
doubled = list(map(lambda x: x * 2, numbers))

# filter() - Keep only even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))

# reduce() - Sum all numbers
sum_all = reduce(lambda x, y: x + y, numbers)

print(f"Doubled: {doubled}")
print(f"Evens: {evens}")
print(f"Sum: {sum_all}")
```

###4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
prime_num = []
for num in range(2, 201):  # Started from 2, as 1 is not prime
    is_prime = True
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            is_prime = False
            break
    if is_prime:
        prime_num.append(num)

print(prime_num)

```

###6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

gen = powers_of_two(10)

for i in gen:
  print(i)
```

###10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def is_vowel(char):
  char.lower()
  mem = ["a", "e", "i", "o", "u"]
  for i in mem:
    return char.translate({ord(i): None for i in 'aeiou'})

string = "Hello Prachi"
charcs = filter(is_vowel, string)
for i in charcs:
  print(i)
```
